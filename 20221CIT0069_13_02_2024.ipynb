```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.set_title('Multiplke lines')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x), label='cos')
ax.set_title('Multiplke lines')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',color='black')
ax.plot(x,np.cos(x), label='cos',color='violet')
ax.plot(x,np.tan(x), label='tan',color='red')
ax.set_title('Multiplke lines')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x), label='cos',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='cos',linestyle='dashdot')
ax.set_title('Multiplke lines')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```
